```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
box=np.ones([20,20])
box_fltr=np.zeros([100,100])
box_fltr[40:60,40:60]=box

plt.figure(figsize=(25,20))

plt.subplot(331),plt.imshow(box_fltr),plt.title("box_filter")

dft=cv2.dft(np.float32(box_fltr),flags=cv2.DFT_COMPLEX_OUTPUT)
img1=cv2.magnitude(dft[:,:,0],dft[:,:,1])

dftshift=np.fft.fftshift(dft)
img2=cv2.magnitude(dftshift[:,:,0],dftshift[:,:,1])


plt.subplot(334),plt.imshow(dft[:,:,0]),plt.title("dft[0]")

plt.subplot(335),plt.imshow(dft[:,:,1]),plt.title("dft[1]")

plt.subplot(336),plt.imshow(img1),plt.title("freq_rspns_of_box_fltr[without shifting]")

plt.subplot(337),plt.imshow(dftshift[:,:,0]),plt.title("dftshift[0]")

plt.subplot(338),plt.imshow(dftshift[:,:,1]),plt.title("dftshift[1]")

plt.subplot(339),plt.imshow(img2),plt.title("freq_rspns_of_box_fltr[with shift]")

```

```python
img=cv2.imread('peppers.png',cv2.IMREAD_GRAYSCALE)
dft=cv2.dft(np.float32(img),flags=cv2.DFT_COMPLEX_OUTPUT)
dftshift=np.fft.fftshift(dft)
rows,cols=img.shape
crow,ccol=int(rows/2),int(cols/2)

cutoff_frequency =100
filterX = cv2.getGaussianKernel(ksize=cutoff_frequency*4+1,
                               sigma=cutoff_frequency)
filterY = filterX.copy()
filter = np.dot(filterX, filterX.T)

mask=np.zeros((rows,cols,2),np.uint8)
mask[crow-200:crow+201,ccol-200:ccol+201,0]=filter
mask[crow-200:crow+201,ccol-200:ccol+201,1]=filter

fshit=dftshift*mask
ishit=np.fft.ifftshift(fshit)
iImg=cv2.idft(ishit)
iImg2 = iImg;
iImg=cv2.magnitude(iImg[:,:,0],iImg[:,:,1])


plt.figure(figsize=(20,10))
plt.subplot(131), plt.imshow(img, cmap='gray')
plt.title('orignal'), plt.axis('off')
plt.subplot(132),plt.imshow(fshit[:,:,0],cmap='gray')
plt.title('inverse'),plt.axis('off')
```
